```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
```

```python
job_salary_per_country = df.pivot_table(
    index="job_country",
    columns="job_title_short",
    values="salary_year_avg",
    aggfunc="median",
)
top_countries = df["job_country"].value_counts().head(6).index
job_salary_top_countries = job_salary_per_country.loc[top_countries]
job_salary_top_countries = job_salary_top_countries[
    ["Data Analyst", "Data Engineer", "Data Scientist"]
]
job_salary_top_countries
```

```python
job_salary_top_countries.plot(kind="bar")
plt.title("Median Salary by Country & Job Title")
plt.xlabel(None)
plt.ylabel("Median Salary ($USD)")
plt.xticks(rotation=25, ha="right")
plt.legend()
plt.show()
```
