```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()  # Look up how to convert to DataFrame?
df.info()
df.head()
```

```python
# Cleaning up data
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
df.info()
# Using median to fill NaN value as it's less skewed than mean due to outliers
salary_year_median = df["salary_year_avg"].median()
salary_hour_median = df["salary_hour_avg"].median()
df_cleaned = df.copy()
df_cleaned.fillna({"salary_year_avg": salary_year_median}, inplace=True)
df_cleaned.fillna({"salary_hour_avg": salary_hour_median}, inplace=True)
df_cleaned.info()
```

```python
# Remove duplicate rows
df_cleaned.drop_duplicates(subset=["job_title", "company_name"], inplace=True)
df_cleaned.info()
```
